```python
#Juni 2019-Agustus 2020
import pandas as pd
import os

# Path ke folder CSV files
folder_path = "/content/drive/MyDrive/menyala ðŸ”¥ðŸ”¥/dataset/older"

# List semua files di folder
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.csv')]

# List buat nyimpen data dari masing-masing CSV
dataframes = []

# Baca masing-masing CSV dan simpan ke list dataframes
for file in all_files:
    try:
        df = pd.read_csv(file)
        if df.empty:
            print(f"Skipping empty file: {file}")
            continue
        dataframes.append(df)
    except pd.errors.EmptyDataError:
        print(f"Skipping empty file: {file}")
        continue

# Gabungin semua dataframes jadi satu
merged_df = pd.concat(dataframes, ignore_index=True)
```

```python
merged_df
```

```python
# Pastikan 'created_at' kolom diubah jadi tipe datetime
merged_df['created_at'] = pd.to_datetime(merged_df['created_at'])

# Ambil kolom 'created_at' dan 'full_text'
selected_columns_df = merged_df[['created_at', 'full_text']]
# Rename
selected_columns_df.rename(columns={'full_text': 'Komentar'}, inplace=True)

# Drop baris dengan nilai NaN
selected_columns_df = selected_columns_df.dropna()

# Sort dataframe berdasarkan 'created_at'
sorted_df = selected_columns_df.sort_values(by='created_at')

# Tampilkan beberapa baris pertama untuk verifikasi
sorted_df
```

```python
# Path to save the CSV file on Google Drive
save_path = '/content/drive/MyDrive/menyala ðŸ”¥ðŸ”¥/dataset/all data/qris_juni_aug20.csv'

# Save the DataFrame to CSV
sorted_df.to_csv(save_path, index=False)

print("DataFrame saved to Google Drive successfully!")
```
